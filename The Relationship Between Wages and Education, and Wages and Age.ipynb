#### The Relationship Between Wages and Education, and Wages and Age

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Importing and displaying the .csv data
df = pd.read_csv('SLID.csv',delimiter=',')
```

```python
print(df)
```

```python
# The data containing rows with null values in the 'wages' column, no row
# with null values will be considered. First, I want to have an idea about 
# rows with null values.

df[np.isnan(df['wages'])]
```

```python
# As you can see, there are 3278 rows with null values in the 'wages' column.
# So, 'wages' being an important column for my analysis, I will take into 
# account only the rows without null values. This can be seen as a sort of data
# cleaning.

df = df[df['wages'] > 1.0]
```

```python
print(df)
```

```python
# As you can see above, 4147 rows contain values which are not null. That is
# the data that I'm going to work with. I will now explore the data using a
# a scatter plot. More specifically, I will show one scatter plot of wages 
# and education, and another of wages and age.

plt.scatter(df['wages'],df['education'])
plt.xlabel('education')
plt.ylabel('wages')
```

```python
plt.scatter(df['wages'],df['age'])
plt.xlabel('age')
plt.ylabel('wages')
```

```python
# Now, I will look at those same relationships using regression.

sns.regplot(df['wages'],df['education'])
```

```python
sns.regplot(df['wages'],df['age'])
```
